1. Uses the data from the Maddison Project to plot the evolution of total population across the world. 

```python
# Let's import pandas and some other basic packages we will use 
from __future__ import division
%pylab --no-import-all
%matplotlib inline
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from IPython.display import Latex
import matplotlib as mpl
import seaborn as sns
# Setup seaborn
sns.set()
from pandas_datareader import data, wb
```

```python
# Setting up paths
import os

pathout = './data/'

if not os.path.exists(pathout):
    os.mkdir(pathout)
    
pathgraphs = './graphs/'
if not os.path.exists(pathgraphs):
    os.mkdir(pathgraphs)
```

```python
# Import data of the Maddison project
if not os.path.exists(pathout + 'Maddison_original.xls'):
    import urllib
    dataurl = "http://www.ggdc.net/maddison/Historical_Statistics/horizontal-file_02-2010.xls"
    urllib.request.urlretrieve(dataurl, pathout + 'Maddison_original.xls')
```

```python
maddison_old_pop = pd.read_excel(pathout + 'Maddison_original.xls', sheet_name="Population", skiprows=2)
maddison_old_pop
```

```python
# Clean and rearrange data
maddison_old_pop.rename(columns={'Unnamed: 0':'Country'}, inplace=True)
```

```python
maddison_old_pop = maddison_old_pop[[col for col in maddison_old_pop.columns if str(col).startswith('Unnamed')==False]]
```

```python
maddison_old_pop.columns = ['Country'] + ['pop_'+str(col) for col in maddison_old_pop.columns[1:]]
```

```python
pop= maddison_old_pop
```

```python
world_popw = pop[pop['Country']=='World Total'] 
```

```python
world_pop = pd.wide_to_long(world_popw, ['pop_'], i='Country', j='year').reset_index()
world_pop = world_pop[world_pop['year']<2030] 
world_pop = world_pop[world_pop['pop_']>0] 
world_pop.dropna(axis=0,subset=['pop_'])
world_pop
```

```python
world_pop = world_pop.pivot_table(index='year',columns='Country',values='pop_',aggfunc='sum')
world_pop
```

```python
# Plot total world population
fig, ax = plt.subplots(figsize=(30,20))
world_pop.plot(ax=ax, linewidth=8)
ax.tick_params(axis = 'both', which = 'major', labelsize=32)
ax.tick_params(axis = 'both', which = 'minor', labelsize=16)
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.legend(prop={'size': 40})
# Label axes
ax.set_xlabel('Year', fontsize=36)
ax.set_ylabel('Population', fontsize=36)
```

2. Plots the evolution of the share of the world population by countries and WB regions.

```python
# Open information of world bank regions in order to merge with country data of Maddison project
wbcountries = wb.get_countries()
wbcountries['name'] = wbcountries.name.str.strip()
wbcountries
```

```python
# Prepare information of both data bases to merge correctly
wbcountries = wbcountries[~wbcountries['region'].str.contains("Aggregates", na = False)]
```

```python
pop2 = pop[~pop['Country'].str.contains("Total", na = False)]
```

```python
pop2 = pop2[pop2['pop_2030'].notna()]
```

```python
pop2['Country'] = pop2['Country'].str.replace(" ","")
```

```python
wbcountries['name'] = wbcountries['name'].str.replace(" ","")
```

```python
pop_WB = pop2.merge(wbcountries, how='left', left_on='Country', right_on='name')
pop_WB
```

```python
pop_WB.at[84,'region']= "East Asia & Pacific"
pop_WB.at[94,'region']= "East Asia & Pacific"
pop_WB.at[81,'region']= "East Asia & Pacific"
pop_WB.at[92,'region']= "East Asia & Pacific"
pop_WB.at[76,'region']= "East Asia & Pacific"
pop_WB.at[58,'region']="Latin America & Caribbean " 
pop_WB.at[66,'region']="Latin America & Caribbean " 
pop_WB.at[125,'region']="Middle East & North Africa" 
pop_WB.at[97,'region']="Middle East & North Africa" 
pop_WB.at[79,'region']="Middle East & North Africa" 
pop_WB.at[106,'region']="Middle East & North Africa" 
pop_WB.at[27,'region']="Europe & Central Asia" 
pop_WB.at[29,'region']="Europe & Central Asia" 
pop_WB.at[31,'region']="Europe & Central Asia" 
pop_WB.at[34,'region']="Europe & Central Asia" 
pop_WB.at[42,'region']="Europe & Central Asia" 
```

```python
noregion = pop_WB[pop_WB['region'].isnull()]
```

```python
# This group of 40 countries is omitted from the final plot. In order to include them it is necessary to assign them a WB region.
noregion
```

```python
pop_WB = pop_WB.pivot_table([col for col in pop_WB if col.startswith('pop_')], 'region', aggfunc='sum')
```

```python
pop_WB = pop_WB.dropna(axis=1, how='any')
```

```python
pop_WB.reset_index(level=0, inplace=True)
```

```python
pop_WB = pd.wide_to_long(pop_WB, ['pop_'], i='region', j='year').reset_index()
```

```python
pop_WB = pop_WB.pivot_table(index='year',columns='region',values='pop_',aggfunc='sum')
pop_WB
```

```python
pop_WB['Total'] = pop_WB.sum(axis=1)
```

```python
regions = ["East Asia & Pacific", "Europe & Central Asia", "Latin America & Caribbean ", "Middle East & North Africa", "North America", "South Asia", "Sub-Saharan Africa "]
```

```python
# Generate a variable of the participation of each world bank region in total population.
for col in regions:
    pop_WB[col] = (pop_WB[col]/pop_WB['Total'])*100
```

```python
pop_WB = pop_WB[pop_WB.columns.drop(list(pop_WB.filter(regex='Total')))]
```

```python
pop_WB2 = pop_WB.loc[ [1, 1000, 1500, 1600, 1700, 1820, 1900, 2000] , : ]
```

```python
# Set the size of the figure and get a figure and axis object
fig, ax = plt.subplots(figsize=(30,20))
# Plot using the axis ax and colormap my_cmap
pop_WB2.loc[1:].plot(ax=ax, linewidth=8)
# Change options of axes, legend
ax.tick_params(axis = 'both', which = 'major', labelsize=32)
ax.tick_params(axis = 'both', which = 'minor', labelsize=16)
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.legend(prop={'size': 22})
# Label axes
ax.set_xlabel('Year', fontsize=36)
ax.set_ylabel('Population (% of total)', fontsize=36)
```

3. Downloads fertility, mortality and life expectancy data from the WB and plots its evolution in the last 60 years.

```python
# Search for variables
fertility = wb.search(string='fertility')
fertility
```

```python
mortality = wb.search(string='mortality')
mortality
```

```python
expectancy= wb.search(string='life expectancy')
expectancy
```

```python
dataid = ['SP.DYN.TFRT.IN', 'SP.DYN.AMRT.MA', 'SP.DYN.AMRT.FE', 'SP.DYN.LE00.IN']
```

```python
wdi = wb.download(indicator=dataid, country=wbcountries.iso2c.values, start=1970, end=2017)
wdi
```

```python
wdi = wdi.groupby(['country', 'year']).max()
wdi.reset_index(inplace=True)
wdi
```

```python
wdi2 = wbcountries.merge(wdi, left_on='name', right_on='country')
wdi2
```

```python
wdi2.rename(columns={'SP.DYN.TFRT.IN':'Fertility rate'}, inplace=True)
```

```python
wdi2.rename(columns={'SP.DYN.AMRT.MA':'Mortality rate male'}, inplace=True)
```

```python
wdi2.rename(columns={'SP.DYN.AMRT.FE':'Mortality rate female'}, inplace=True)
```

```python
wdi2.rename(columns={'SP.DYN.LE00.IN':'Life expectancy'}, inplace=True)
```

```python
wdi2
```

```python
wdi_life = wdi2.pivot_table(index='year', values='Life expectancy',aggfunc='mean')
```

```python
wdi_life
```

```python
# Plot life expectancy
fig, ax = plt.subplots(figsize=(30,20))
wdi_life.plot(ax=ax, linewidth=8)
ax.tick_params(axis = 'both', which = 'major', labelsize=32)
ax.tick_params(axis = 'both', which = 'minor', labelsize=16)
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.legend(prop={'size': 40})
# Label axes
ax.set_xlabel('Year', fontsize=36)
ax.set_ylabel('Average life expectancy', fontsize=36)
```

```python
wdi_mortM = wdi2.pivot_table(index='year', values='Mortality rate male',aggfunc='mean')
```

```python
# Plot infant mortality
fig, ax = plt.subplots(figsize=(30,20))
wdi_mortM.plot(ax=ax, linewidth=8)
ax.tick_params(axis = 'both', which = 'major', labelsize=32)
ax.tick_params(axis = 'both', which = 'minor', labelsize=16)
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
ax.legend(prop={'size': 40})
# Label axes
ax.set_xlabel('Year', fontsize=36)
ax.set_ylabel('Average infant mortality rate', fontsize=36)
```
